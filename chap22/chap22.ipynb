```python
from transformers import pipeline

classifier = pipeline(task="sentiment-analysis", device='cuda')
preds = classifier("The Los Angeles Chargers won the game, and I'm extremely happy!")
print(preds)
```

```python
classifier = pipeline(model="ProsusAI/finbert", task="sentiment-analysis", device='cuda')
preds = classifier("This is the worst-tasting burger I've ever had.")
print(preds)
```

```python
translation = pipeline(model="t5-base", task="translation", device='cuda')
result = translation("translate English to German: The house is wonderful.")
print(result)

```

```python
import os, requests, json

API_TOKEN = os.environ.get("HUGGINGFACE_API_KEY")

model = "google/flan-t5-xxl"
API_URL = f"https://api-inference.huggingface.co/models/{model}"
headers = {"Authorization": f"Bearer {API_TOKEN}", "Content-Type": "application/json"}

def query(payload, api_url=API_URL, headers=headers):
    data = json.dumps(payload)
    response = requests.request("POST", api_url, headers=headers, data=data)
    return json.loads(response.content.decode("utf-8"))

question = "Which region is known for producing French red wine?"
data = query({"inputs" : question})

print(data)
```

```python
model = "intfloat/multilingual-e5-large"
API_URL = f"https://api-inference.huggingface.co/models/{model}"

question = "Today is a cold and overcast day, I hope the sun comes out tomorrow."
data = query({"inputs" : question}, api_url=API_URL)

print(data)
```
